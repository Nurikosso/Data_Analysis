```python
df = pd.read_csv(r"C:\Users\nurik\Desktop\курс Python\Global Superstore Data\Data_Analysis\NoteBooks\Cleaned_data.csv")
df
```

```python
# Уствноваил столбец "Row.ID" как индекс столбцов

df = df.set_index('Row.ID')
```

```python
# Исходя из инпута данного кода видно, что страна с большим профитом это
# Соединенноые Штаты. Так же в топ 10 Стран вошли Япония, Италия и Индия


df_profit = df.sort_values('Profit', ascending=False)
df_profit.head(10)

# Строка "df/sort_values('Profit', ascending=False)" сортирует данные в столбце Profit по убыванию, с большей к меньше, а "df_profit =" присаивает этот результат новому датафрейму
```

```python
# Данный аутпут показывает мне, что самым прибыльным рынком в таблице является потребительский сегмент
# а самым менее прибыльным домашний офис

df_seg = df.groupby('Segment')['Profit'].sum()
df_seg

# строка "df.groupby('Segment')['Profit']" группирует данные из столбца "Profit" в соотношении столбца "Segment", а ".sum()" 
# показывает мне сумму данных профита по каждой группе. "df_seg =" присваивает данные новому DataSeries.
```

```python
# В данном аутпуте вижу, что самое большое количество продаж приходит на категорию "Technology"
# а самое меньшее на категорию "Office Supplies".

df_sales = df.groupby('Category')['Sales'].sum().sort_values()
df_sales

# Код "df.groupby('Category')['Sales'].sum()" группирует и суммирует продажи из колонки "Sales" относительно каждой
# категории товаров, "sort_values()" соритирует от меньшей, к большей
```

```python
# В данном аутпуте видно, что самым продаваемым товаром являются телефоны и принтеры в категории технологий
# а самым непродаваемым стали ярлыки(закладки) в категории офисных товаров. Также видно, что из мебельной продукции
# больше всего покупок среди кресел и книжных шкафов, а среди офисных товаров это разного рода хранилища. В целом по данным также подтверждается
# идея о том, что офисные товары менее востребованы

df_sales2 = df.groupby(['Category', 'Sub.Category'])['Sales'].sum().sort_values(ascending=False)
df_sales2

# Строка кода "df.groupby(['Category', 'Sub.Category'])['Sales'].sum(" отгруппировала для меня данные по продажам относительно каждой категории, а потом подкатегории
# и привела мне сумму продаж, а ".sort_values(ascending=False)" отсортировала продажи от большиъ к меньшим
```

```python
# Исходя из данных этого аутпута видно, что несмотря на то, что количество продаж в категории 'Office Supplies' меньше, чистый профит там оказался больше
# чем в категории "Furniture". Это указывает нам о том, что маржинальность в категории офисной принадлежности больше, чем в категории мебели

df_prof_group = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
df_prof_group
```

```python
# Данный фрейм был призван доказать еще раз, что маржинальность каждой единицы товара в офисных принадлежностях больше, чем в категории мебель,
# однако он также указал нам, что при большем общем отрыве категории технологий, в разрезе маржинальность каждой единицы товара на количество продаж, офисные принадлежности
# всего немного уступают товарам технологии. Это еще больше указывает о том, что офисные товары, вероятно, имеют больший ценник, либо меньшую себестоимость

df_avg_prof = (df.groupby('Category')['Profit'].sum() / df.groupby('Category')['Sales'].sum())
df_avg_prof

# Строка "df.groupby('Category')['Profit'].sum()" группирует сумму профита среди всех категорий, "df.groupby('Category')['Sales'].sum()" группирует сумму продаж в каждой категории
# а в общем код "df.groupby('Category')['Profit'].sum() / df.groupby('Category')['Sales'].sum()" делит сумму одного датафрейма на сумму другого
```

```python
# На данном результате видно превосходство принтеров среди всех подкатегорий, несмотря на большее количество продаж телефонов (0.054496 у принтеров против 0.045676 у телефонов). Связываю это с большей себестоимостью производства
# телефона. Также наблюдается отрицательный показатель маржинальности столов в категории мебели, что вероятно и сказывается на общем отставании по показателям среди остальных категорий в купе с меньшими продажами.
# Видно, что отрицательная маржинальность столов является исключительным случаем. Возможно, это связано со скидками, низким спросом, или дорогой доставкой.

df_avg_prof2 = (df.groupby(['Category', 'Sub.Category'])['Profit'].sum() / df.groupby('Category')['Sales'].sum()).sort_values(ascending=False)
df_avg_prof2
```

```python
# При изучении соотношения скидок в каждый категории и подкатегории видим, что столы в категории "Furiture" продавались со значительно большей скидкой, чем остальные, что, вероятно, и привело к отрицательной маржинальности данной 
# подкатегории. Исходя из общей информации таблицы видно, что в целом офисные принадлежности продавались с меньшими скидками, чем товары категории мебель, исходя из чего и понятно, по какой причине офисные принадлежности при меньших
# продажах принесли больше общей прибыли, и прибыли на единицу товара. Также можно заметить, что на телефоны скидки были больше, чем на принтеры, исходя из чего и делается вывод, что маржинальность на единицу товара у принтера будет
# больше, чем у телефона, несмотря на большее количество продаж в сумме у телефонов (примерно на 200 тысяч)

df_disc = df.groupby(['Category', 'Sub.Category'])['Discount'].mean().sort_values(ascending=False)
df_disc
```

```python
# В данном выводе вижу, что скидка отрицательно влияет на профит, однако связь не такая уж и сильная. Чем больше скидка, тем меньше маржа товаров. Исходя из данных делается вывод, что на маржинальность влияют и другие факторы
# например, себестоимость товара, спрос на него. Скидки могут быть эффективным инструментом для поднятия продаж, однако стоит учитывать их влияние на прибыльность с каждой продажи. 

corr = df[['Discount', 'Profit']].corr()
corr
```

```python
# Из данного аутпута видно, что скидки чаще влияют отрицательно на количество
# продаж. Особенно это видно в категории "Technology". Судя по всему, покупатели
# этой категории ориентируются на качество товара, новизну, бренд, и другие
# показатели. Видно, что скидки положительно влияют на категорию мебели
# однако это влияние слабо выражено, что также подтверждает неэффективность.
# Самый отрицательный показатель оказался в категории "Office Supplies" и 
# подкатегории "Art" с показателем "-0.071567" Самые высокие результаты
# показала категория также "Office Supplies", но уже "Supplies" с показателем
# "0.025759", что в общей картине также немного.


df_cor2 = df.groupby(['Category', 'Sub.Category'])[['Quantity', 'Discount']].corr()
df_cor2
```

```python
# В данном аутпуте видно, что при разных классах доставки профит не растет
# практически. Однако доставка первым классом принесла меньше прибыли, и
# имеет большую среднюю скидку. Это может указывать, что покупатели могут
# выбирать первый класс при меньшей стоимости самого товара, экономя за счет
# скидок. Доставки стандартным классом принесли больше всех профита при чуть ли не
# худших показателях количества продаж. Вероятно, это связано с количеством
# "Quantity", покупок в чеке.

ship_disc = df.groupby('Ship.Mode')[['Discount', 'Profit', 'Sales']].mean()
ship_disc
```

```python
# В данном коде подтверждается идея о том, что покупатели стандарт класса
# приносят больше прибыли из-за количества покупок в чеке

df.groupby('Ship.Mode')['Quantity'].mean()
```

```python
# В данном ответе от программы подтверждается, что основным рынком сбыта
# является сегмент потребителей со стандартным типом доставки (15876 раз, что
# гораздо больше, чем у любого другого сегмента, или способа доставки). Также
# видно что доставка день в день не является важным критерием для корпоративных
# покупателей и для покупателей в домашний офис. Однако для потребителей
# доставка день в день пользуется большим спросом (1467), чем доставка первым
# классом для покупок в домашний офис или корпоративных покупателей (506 и 728)
# Доставка день в день или первым классом, вероятнее всего, менее востребована
# у корпоративных покупателей, так как заказы планируются заранее.

df.groupby(['Segment', 'Ship.Mode'])['Sales'].count().sort_values(ascending=False)
```

```python
# Проверив данный ответ, стало ясно, какие страны приносят большую прибыль
# а какие приносят больше убытков, чем прибыли. Видно, что в тройке лидеров у нас
# по большей части страны регоина EMEA, и в тройке самых непрофитных регионов
# также они. Это свидетельствует, что профит по продажам больше зависит от
# страны покупателя, нежели от региона.

df_country_profit = df.groupby(['Country', 'Region'])['Profit'].mean()
df_country_profit.sort_values(ascending=False)
```

```python
# При более детальном изучении тройки худших стран по прибыльности видно, что
# дело может быть в скидках, которые они получают при покупках

filt_lit = df[df['Country'].isin(['Lithuania', 'Turkmenistan', 'Portugal'])]
filt_lit.sort_values(by='Profit', ascending=True)
```

```python
# Данный вывод показывает слабо выраженную зависимость между скидками
# и профитом. Это указывает, что отрицательный профит может возникнуть вследсвтие
# других причин. Возможно, из-за стоимости доставки, или структуры рынка данных
# стран (налоговая пошлина, например)

filt_lit[['Discount', 'Profit']].corr()
```

```python
# Данная таблица наглядно показывает маржинальность в каждом из регионов мира
# В таблице четко видно выгоду продажи в Канаде (0.266201), и нерентабельность 
# продаж в регионах Юго-Восточной Азии (0.020185)

region_margin = df.groupby('Region', group_keys=False).apply(lambda x: x['Profit'].sum() / x['Sales'].sum())
region_margin.sort_values(ascending=False)

# строка ".apply(lambda x: x['Profit'].sum() / x['Sales'].sum())" применяет функцию к каждой строке,
# группируя их по регионам
```

```python
# В данном аутпуте видно, что больше всего заказов приходится на Соединенные
# Штаты, при этом в Канаде заказов не так много (201). С учетом большой общей
# маржинальности Канады, это указывает, что покупатели оттуда берут более
# дорогие товары, нежели Соединенные Штаты. При этом меньше всего заказов было

df.groupby(['Country'])['Order.ID'].nunique().sort_values(ascending=False)
```

```python
# В данном аутпуте видно, что самые маленькие заказы были в странах EMEA
# и Африки, что может и снижать прибыльность этих регионов для продавца
# с учетом стоимости доставки, налоговых пошлин, и других факторов

df.groupby(['Country', 'Region'])['Order.ID'].nunique().sort_values()
```
