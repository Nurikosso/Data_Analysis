```python
import pandas as pd
```

```python
# В данном коде указываю, что необходимо показать мне все столбцы

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(r"C:\Users\nurik\Desktop\курс Python\Global Superstore Data\Data_Analysis\NoteBooks\Cleaned_data.csv")
df
```

```python
# Уствноваил столбец "Row.ID" как индекс таблицов

df = df.set_index('Row.ID')
```

```python
# Исходя из инпута данного кода видно, что страна с большим профитом это
# Соединенноые Штаты. Так же в топ 10 Стран вошли Япония, Италия и Индия


df_profit = df.sort_values('Profit', ascending=False)
df_profit.head(10)

# Строка "df/sort_values('Profit', ascending=False)" сортирует данные в столбце Profit по убыванию, с большей к меньше, а "df_profit =" присаивает этот результат новому датафрейму
```

```python
# Данный аутпут показывает мне, что самым прибыльным рынком в таблице является потребительский сегмент
# а самым менее прибыльным домашний офис

df_seg = df.groupby('Segment')['Profit'].sum()
df_seg

# строка "df.groupby('Segment')['Profit']" группирует данные из столбца "Profit" в соотношении столбца "Segment", а ".sum()" 
# показывает мне сумму данных профита по каждой группе. "df_seg =" присваивает данные новому DataSeries.
```

```python
# В данном аутпуте вижу, что самая большое количество продаж приходит на категорию "Technology"
# а самое меньшее на категорию "Office Supplies".

df_sales = df.groupby('Category')['Sales'].sum().sort_values()
df_sales

# Код "df.groupby('Category')['Sales'].sum()" группирует и суммирует продажи из колонки "Sales" относительно каждой
# категории товаров, "sort_values()" соритирует от меньшей, к большей
```

```python
# В данном аутпуте видно, что самым продаваемым товаром являются телефоны и принтеры в категории технологий
# а самым непродаваемым стали ярлыки(закладки) в категории офсиных товаров. Также видно, что из мебельной продукции
# больше всего покупок среди кресел и книжных шкафов, а среди офисных товаров это рахного рода хранилища. В целом по данным также подтверждается
# идея о том, что офисные товары менее востребованы

df_sales2 = df.groupby(['Category', 'Sub.Category'])['Sales'].sum().sort_values(ascending=False)
df_sales2

# Строка кода "df.groupby(['Category', 'Sub.Category'])['Sales'].sum(" отгруппировала для меня данные по продажам относительно каждой категории, а потом подкатегории
# и привела мне сумму продаж, а ".sort_values(ascending=False)" отсортировала продажи от большиъ к меньшим
```

```python
# Исходя из данных этого аутпута видно, что несмотря на то, что количество продаж в категории 'Office Supplies' меньше, чистый профит там оказался больше
# чем в категории "Furniture". Это указыавет нам о том, что маржинальность в категории офисной принадлежности больше, чем в категории мебели

df_prof_group = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
df_prof_group
```

```python
# Данный фрейм был призван доказать еще раз, что маржинальность каждой единицы товара в офисных принадлежностях больше, чем в категории мебель,
# однако он также указал нам, что при большем общем отрыве категории технологий, в разрезе маржинальность каждой единицы товара на количество продаж, офисные принадлежности
# всего немного уступают товарам технологии. Это еще больше указывает о том, что офисные товары, вероятно, имеют больший ценник, либо меньшую себестоимость

df_avg_prof = (df.groupby('Category')['Profit'].sum() / df.groupby('Category')['Sales'].sum())
df_avg_prof

# Строка "df.groupby('Category')['Profit'].sum()" группирует сумму профита среди всех категорий, "df.groupby('Category')['Sales'].sum()" группирует сумму продаж в каждой категории
# а в общем код "df.groupby('Category')['Profit'].sum() / df.groupby('Category')['Sales'].sum()" делит сумму одного датафрейма на сумму другого
```

```python
# На данном результате видно превосходство принтеров среди всех подкатегорий, не смотря на большее количество продаж телефонов (0.054496 у принтеров против 0.045676 у телефонов). Свзяываю это с большей себестоимостью производства
# телефона. Также наблюдается отрицательный показатель маржинальности столов в категории мебели, что вероятно и сказывается на общем отставании по показателям среди остальных категорий в купе с меьшими продажами.
# Видно, что отрицательная маржинальность столов является исключительным случаем. Возможно, это связано со скидками, низким спросом, или дорогой доставкой.

df_avg_prof2 = (df.groupby(['Category', 'Sub.Category'])['Profit'].sum() / df.groupby('Category')['Sales'].sum()).sort_values(ascending=False)
df_avg_prof2
```

```python
# При изучении соотношения скидок у каждый категории и подкатегории видим, что столы в категории "Furiture" продавались со значительно большей скидкой, чем остальные, что, вероятно, и привело к отрицательной маржинальности данной 
# подкатегории. Исходя из общей информации таблицы видно, что в целом офсиные принадлежности продавались с меньшими скидками, чем товары кагеории мебель, исходя из чего и понятно, по какой причине офисные принадлежности при меньших
# продажах принесло больше общей прибыли, и прибыли на единицу товара. Также можно заметить, что на телефоны скидки были больше, чем на принтеры, исходя из чего и делается вывод, что маржинальность на единицу товара у принтера будет
# больше, чем у телефона, несмотря на большее количество продаж в сумме у телефонов (приерно на 200 тысяч)

df_disc = df.groupby(['Category', 'Sub.Category'])['Discount'].mean().sort_values(ascending=False)
df_disc
```

```python
# В данном выводе вижу, что скидка отрицательно влияет на профит, однако связь не такая уж и сильная. Чем больше скидка, тем меньше маржа товаров. Исходя из данных делается вывод, что на маржинальность влияют и другие факторы
# например, себестоимость товара, спрос на него. Скидки могут быть эффективным инструментом для поднятия продаж, однако стоит учитывать их влияние на прибыльность с каждой продажи. 

corr = df[['Discount', 'Profit']].corr()
corr
```
