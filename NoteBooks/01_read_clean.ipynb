```python
# Проверка уникальных значений в каждом столбце.
# Просматривает информация, например, что информация собрана за 4 года
# Также, видно, что продажи проводились на 7 маркетах в 13 регионах мира
# Всего имеется 4 класса доставки и 3 сегмента покупателя с 4 уровнями
# приоритетности. Нашел странную 53 неделю в году, проверяю

df.nunique()
```

```python
# Проверяю количество уникальных значений в столбце "weeknum"
# В году их должно быть 52
df['weeknum'].unique()
```

```python
# После вывода всех 53 недель заказов, обнаружил, что на самом деле
# все корректно, в 2012 году 31 декабря стало еще одной неделей декабря (ПН)
# что перенесло даты до 6 января в прошлый год пятую неделю

df.loc[df['weeknum'] == 53]
```

```python
# В данном блоке кода установил столбец "Row.ID" как индекс
# а также сделал сортировку от первого до последнего
# Это сделает таблицу более приятной и воспринимаемой для глаз

df = df.set_index('Row.ID')
df = df.sort_index()
df
```

Перехожу к более детальному изучению столбцов и строк в данном датасете, изучаю, какие строки или столбцы можно удалить за ненадобностью, чтобы разгрузить таблицу


```python
# При обследовании таблицы через ".describe" было обнаружено, что столбец (记录数)
# имеет всего одно значение "1", проверяю сортировкой, подтверждается

df.sort_values(df.columns[7], ascending=False)
```

```python
# Также решил проверить количество уникальных записей в данном столбце, для 
# более точного подтверждения, чтобы не осталось сомнений о бесполезности
# данного столбца

df[df.columns[7]].unique()
```

```python
# В данном коде избавился от лишнего столбца (记录数), который означал "Количество записей"
# Столбец бесполезный, ведь каждая строка уникальна, нет ни одной строки,
# полностью повторяющейся. 

df = df.drop(df.columns[7], axis=1)
df
```
